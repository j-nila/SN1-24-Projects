# Welcome to Blackjack Simulator
The goal of blackjack is simple.  To win, one must have a higher hand value than the dealer, without going over 21. Players are dealt two cards and can then choose to `hit` (receive additional cards), `double and hit` (receive an additional card, double the bet they placed down, and end their turn immediately), or `stand` (keep their current hand). The dealer also receives two cards, but only one is face up.

If a player’s hand exceeds 21, they “bust” and lose the game. If the dealer busts, all remaining players win. If neither the player nor the dealer busts, the player with the highest hand value wins (keep their original bet and the dealer matches it). If the player and the dealer have the same total, nobody wins. If the player gets a total of 21 from their first 2 cards (a Blackjack), they get their original bet back and also win 1.5 the bet they placed initially, unless the dealer also gets a Blackjack, in which case nobody wins.

A player may bet between 10 to 150$ per round. No cents allowed. If a player has no more money to bet, the game ends and each player keeps the money they currently have.
